```python
# to open the excel sheet we need to install a package called openpyxl
# to install type "pip install openpyxl" in terminal

from openpyxl import load_workbook

# keep the data file and mpthon code in same directory
wb = load_workbook('dataa.xlsx', data_only=True)
sheet = wb['sample data(191001~191002)']


q = 0


while(q == 0):
    print("choose for the menu:")
    print("1. U_i,t")
    print("2. Y_j,t")
    print("3. X_i,j,20191001")
    print("4. X_i,j,20191002")
    print("5. quit")
    val = int(input("enter your choice : "))

    # part where U_i,t is calculated
    # subject regarding capacity(2).
    if val == 1:
        sum3 = 0
        d = 0
        i = input('enter value of i: ')
        t = input('enter value of t: ')
        for j in range(3, 17):

            if sheet.cell(row=j, column=3).value == i and sheet.cell(row=j, column=4).value == t:
                sum3 = sum3 + int(sheet.cell(row=j, column=11).value)
                d = d+int(sheet.cell(row=j, column=17).value)

        # checking the subject 5 U is always greater than equal to 0 U_i,t >=0
        if d-sum3 >= 0:
            u = d-sum3
        else:
            u = 0

        print('U_i,t is {}'.format(u))

    # part where Y_j,t is calculated
    elif val == 2:
        flag = 1
        j = input('enter value of j: ')
        t = input('enter value of t: ')
        for i in range(3, 17):
            if sheet.cell(row=i, column=1).value == j and sheet.cell(row=i, column=4).value == t:
                print(1)
                flag = 1
                break
            else:
                flag = 0

        if flag == 0:
            print(0)

    # part where X_i,j,t is calcluated for t=20191001
    elif val == 3:
        sum1 = 0
        j = input('enter value of j: ')
        t = input('enter value of i: ')
        y = str(20191001)
        for i in range(3, 17):

            if sheet.cell(row=i, column=1).value == j and sheet.cell(row=i, column=3).value == t and sheet.cell(row=i, column=4).value == y:
                sum1 = sum1 + int(sheet.cell(row=i, column=11).value)

        # checking subject 1,3,4
        if sum1 <= int(sheet.cell(row=i, column=16).value) and sum1 >= 0:
            print(' X_i,j,20191001 is {}'.format(sum1))

    # part where X_i,j,t is calcluated for t=2019002
    elif val == 4:
        sum2 = 0
        j = input('enter value of j: ')
        t = input('enter value of i: ')
        y = str(20191002)
        for i in range(3, 17):

            if sheet.cell(row=i, column=1).value == j and sheet.cell(row=i, column=3).value == t and sheet.cell(row=i, column=4).value == y:
                sum2 = sum2 + int(sheet.cell(row=i, column=11).value)

        # checking subject 1,3,4
        if sum2 <= int(sheet.cell(row=i, column=16).value) and sum2 >= 0:
            print(' X_i,j,20191002 is {}'.format(sum2))

    elif val == 5:
        q = 1

    else:
        print('wrong option')

    # to calcuate the value of minize z
    set_i = ['059396', '061724', '066130',
             '057835', '057713', '063843', '069569']
    set_t = ['20191001', '20191002']
    z = 0

    for m in set_i:
        for n in set_t:
            sum4 = 0
            d = 0
            u = 0

            for j in range(3, 17):

                if sheet.cell(row=j, column=3).value == m and sheet.cell(row=j, column=4).value == n:
                    sum4 = sum4 + int(sheet.cell(row=j, column=11).value)
                    d = d+int(sheet.cell(row=j, column=17).value)
            if d-sum4 >= 0:
                u = d-sum4

            z = z+u

    print('the minimize z is {}'.format(z))
```
